```python
import sys
# install httpimport if not in current python environment
!{sys.executable} -m pip install httpimport

# import httpimport - this allows you to import remote
# modules from urls, like github
import httpimport

httpimport.INSECURE = True

# Import heatunits script / module from github
with httpimport.remote_repo('https://github.com/UCANR-IGIS/caladapt-py/tree/master/CalAdaptLib'):
    import HeatUnits as hu

# Import pandas
import pandas as pd
```

```python
# CSV from degree day challenge
csv = r'C:\Users\stfei\Desktop\espartoa-weather-2020.csv'

# Read CSV into pandas dataframe
data = pd.read_csv(csv)

# add column for yesterdays min temp
data["tmin_tom"] = data["tmin"].shift(-1)
data.tmin_tom.fillna(data.tmin, inplace=True)

# add column for tomorrows max temp
data["tmax_yest"] = data["tmax"].shift(1)
data.tmax_yest.fillna(data.tmax, inplace=True)

# print data
data
```

```python
# 1 = Hor. or 2 = Int. or 3 = Vert. Cutoff

# Apply each of the Degree Day Calculations using the heatunits script/module
data["SiSineHor"] = data.apply(lambda x: hu.HeatU(50, 70, 1, 1, 1, x['tmin'], x['tmax']), axis=1)
data["SiSineInt"] = data.apply(lambda x: hu.HeatU(50, 70, 1, 2, 1, x['tmin'], x['tmax']), axis=1)
data["SiSineVert"] = data.apply(lambda x: hu.HeatU(50, 70, 1, 3, 1, x['tmin'], x['tmax']), axis=1)

data["DoSineHor"] = data.apply(lambda x: hu.HeatU(50, 70, 2, 1, 1, x['tmin'], x['tmax'], x['tmin_tom'], x['tmax_yest']), axis=1)
data["DoSineInt"] = data.apply(lambda x: hu.HeatU(50, 70, 2, 2, 1, x['tmin'], x['tmax'], x['tmin_tom'], x['tmax_yest']), axis=1)
data["DoSineVert"] = data.apply(lambda x: hu.HeatU(50, 70, 2, 3, 1, x['tmin'], x['tmax'], x['tmin_tom'], x['tmax_yest']), axis=1)

data["SiTriHor"] = data.apply(lambda x: hu.HeatU(50, 70, 3, 1, 1, x['tmin'], x['tmax']), axis=1)
data["SiTriInt"] = data.apply(lambda x: hu.HeatU(50, 70, 3, 2, 1, x['tmin'], x['tmax']), axis=1)
data["SiTriVert"] = data.apply(lambda x: hu.HeatU(50, 70, 3, 3, 1, x['tmin'], x['tmax']), axis=1)

data["DoTriHor"] = data.apply(lambda x: hu.HeatU(50, 70, 4, 1, 1, x['tmin'], x['tmax'], x['tmin_tom'], x['tmax_yest']), axis=1)
data["DoTriInt"] = data.apply(lambda x: hu.HeatU(50, 70, 4, 2, 1, x['tmin'], x['tmax'], x['tmin_tom'], x['tmax_yest']), axis=1)
data["DoTriVert"] = data.apply(lambda x: hu.HeatU(50, 70, 4, 3, 1, x['tmin'], x['tmax'], x['tmin_tom'], x['tmax_yest']), axis=1)

data = data.round({'SiSineHor': 2, 'SiSineInt': 2, 'SiSineVert': 2,
                   'DoSineHor': 2, 'DoSineInt': 2, 'DoSineVert': 2,
                   'SiTriHor': 2, 'SiTriInt': 2, 'SiTriVert': 2,
                   'DoTriHor': 2, 'DoTriInt': 2, 'DoTriVert': 2})

# Print the resulting dataframe with each calculation as a column
data
```
